```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Data historis
data = {
    'Tahun': [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022],
    'Pendaftar USM': [400, 436, 600, 700, 532, 349, 926, 1410, 1326, 1234, 1065, 1092, 1184, 842, 774, 885, 795, 437, 638, 561, 586, 507, 582],
    'Lulus USM': [309, 316, 345, 381, 452, 320, 574, 872, 864, 847, 829, 690, 799, 540, 491, 598, 584, 308, 513, 503, 532, 453, 447],
    'Registrasi Ulang': [309, 312, 329, 365, 443, 280, 502, 647, 688, 722, 645, 651, 760, 511, 467, 572, 567, 297, 453, 406, 401, 335, 217]
}

# Mengubah data ke dalam DataFrame
df = pd.DataFrame(data)

# Menyusun perubahan jumlah mahasiswa baru
df['Perubahan Pendaftar'] = df['Pendaftar USM'].pct_change().fillna(0)  # Persentase perubahan tahun ke tahun
mean_change = df['Perubahan Pendaftar'].mean()  # Rata-rata perubahan
std_dev_change = df['Perubahan Pendaftar'].std()  # Standar deviasi perubahan

# Simulasi Monte Carlo
num_simulations = 1000
num_years = 5  # Prediksi selama 5 tahun ke depan

simulations = []

for _ in range(num_simulations):
    last_value = df['Pendaftar USM'].iloc[-1]
    simulated_years = []
    for _ in range(num_years):
        random_change = np.random.normal(mean_change, std_dev_change)  # Perubahan acak berdasarkan distribusi normal
        last_value *= (1 + random_change)  # Menghitung nilai berdasarkan perubahan
        simulated_years.append(last_value)
    simulations.append(simulated_years)

# Menghitung rata-rata dan interval prediksi
simulated_df = pd.DataFrame(simulations).T
mean_prediction = simulated_df.mean(axis=1)
lower_bound = simulated_df.quantile(0.05, axis=1)
upper_bound = simulated_df.quantile(0.95, axis=1)

# Menampilkan hasil simulasi dalam tabel
predicted_years = [2023 + i for i in range(num_years)]
prediction_table = pd.DataFrame({
    'Tahun': predicted_years,
    'Prediksi Mean': mean_prediction,
    'Prediksi 5%': lower_bound,
    'Prediksi 95%': upper_bound
})

print(prediction_table)

# Plotting hasil simulasi
plt.figure(figsize=(10, 6))
plt.plot(predicted_years, mean_prediction, label='Prediksi Mean', color='blue')
plt.fill_between(predicted_years, lower_bound, upper_bound, color='gray', alpha=0.5, label='95% Interval')
plt.title('Simulasi Prediksi Jumlah Mahasiswa Baru (Monte Carlo)')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Pendaftar')
plt.legend()
plt.grid(True)
plt.show()
```
